```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing datset
dataset = pd.read_csv('student_info.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

```python
#taking care of missing data
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer.fit(x[:,:1])
x[:,:1] = imputer.transform(x[:,:1])
```

```python
print(x)
```

```python
#encoding dependent variable which is y
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

```python
#splitting dataset into training set and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 51)
print("shape of x_train =", x_train.shape)
print("shape of y_train =", y_train.shape)
print("shape of x_test =", x_test.shape)
print("shape of y_test =", y_test.shape)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
#training logistic regression model on training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 51)
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
pd.DataFrame(np.c_[x_test,y_test,y_pred],columns=["study_hours","student_marks_original","student_marks_predicted"])
```

```python
#visualizing training set results
plt.scatter(x_train,y_train,color = 'red')
plt.plot(x_train, classifier.predict(x_train), color = 'blue')
plt.title('Study Hours vs Student Marks (Training set)')
plt.xlabel('Study Hours')
plt.ylabel('Student Marks')
plt.show()
```

```python
#visualizing test set results
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, classifier.predict(x_train), color = 'blue')
plt.title('Study Hours vs Student Marks (Test set)')
plt.xlabel('Study Hours')
plt.ylabel('Student Marks')
plt.show()
```
